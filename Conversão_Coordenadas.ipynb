### Separação das coordenadas unidas

> Para a utilização do separador são necessarias duas informações, a primeira a ser apresentada é o banco de coordenadas e o segundo o separador. O banco de dados deve ser apresentado sozinho, caso seja composto de uma única coluna e.g. dados, porém se for apresentado em um dataset contendo mais de uma coluna a coluna deve ser apresentada e.g. dados['unidas']. E o separado é o símbolo que separa as duas partes da coordenada, e.g. ';', ',', e deve ser apresentado entre aspas.

```python
def separador(dados, sep):
    lt = []
    ln = []
    for i in dados:
        x, y = i.split(sep) 
        lt.append(x)
        ln.append(y)

    data['Longitude'] = ln
    data['Latitude'] = lt
    return data
```

```python
# e.g.:
separador(data['unidas'], ',') #Irá gerar duas novas colunas Latitude e Longitude em seu dataset.
```

# Convertendo as coordenadas

> Para a conversão das coordenadas, deverão existir no seu dataset uma coluna denominada Longitude e outra Latitude. Se você utilizou a função separador, essas colunas já existem. Depois de se certificar da existência dessas colunas, com a primeira letra maiúscula, utilize a função conversor( ) para realizar a conversão.

```python
def conversion_i(old):
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    new = old.replace(u'°',' ').replace('\'',' ').replace('"',' ').replace('º',' ').replace('”', ' ').replace('′',' ').replace('″',' ').replace('’',' ')
    new = new.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(float(new[0]))+int(float(new[1]))/60.0+int(float(new[2]))/3600.0) * direction[new_dir]
```

```python
def conversion_p(old):
    direction = {'N':1, 'S':-1, 'L': 1, 'O':-1}
    new = old.replace(u'°',' ').replace('\'',' ').replace('"',' ').replace('º',' ').replace('”', ' ').replace('′',' ').replace('″',' ').replace('’',' ')
    new = new.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(float(new[0]))+int(float(new[1]))/60.0+int(float(new[2]))/3600.0) * direction[new_dir]
```

```python
def conversor(data):
    long = []
    lat = []
    
    for i in data.Longitude:
        try:
            x = conversion_i(i)
            long.append(x)
        except:
            x = conversion_p(i)
            long.append(x)
    
    for i in data.Latitude:
        try:
            x = conversion_i(i)
            lat.append(x)
        except:
            x = conversion_p(i)
            lat.append(x)
    
    data['Long_decimal'] = long
    data['Lat_decimal'] = lat
    
    return data
```

```python
#e.g^.:

data = conversor(data)
data
```
